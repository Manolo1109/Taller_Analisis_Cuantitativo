```python
from sklearn.model_selection import train_test_split
Y = df_punto3_estandarizado_xy["Densidad"]
X = df_punto3_estandarizado_xx
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
correlation_spearman_xy = XY.corr(method="spearman")
correlation_spearman_xy = correlation_spearman_xy.drop(['Densidad'],axis=0)
correlation_spearman_xy = correlation_spearman_xy['Densidad']
correlation_spearman_xy
```

```python
correlation_spearman_xx = X_train.corr(method="spearman")
correlation_spearman_xx
```

```python
inverse_correlation_spearman_xx = np.linalg.inv(correlation_spearman_xx)
inverse_correlation_spearman_xx = pd.DataFrame(inverse_correlation_spearman_xx)
inverse_correlation_spearman_xx
```

```python
# Calculamos los coeficientes del modelo RLM usando la fórmula
beta_spearman = np.dot(inverse_correlation_spearman_xx, correlation_spearman_xy)
beta_spearman = pd.DataFrame(beta_spearman)
beta_spearman
```

```python
B0 = uy - np.dot(np.transpose(ux),beta_spearman)
B0 = pd.DataFrame(B0)
B0 = B0.at[0, 0]
B0
```

```python
Y_pred_spearman = B0 + np.dot(X_test, beta_spearman)
Y_pred_spearman = pd.DataFrame(Y_pred_spearman)
Y_pred_spearman
y_pred_test= Y_pred_spearman.copy()
y_pred_test
```

```python
# Implementamos una función que nos permita calcular las métricas de regresión para el conjunto 
# de entrenamiento y de prueba
# R2, R2 Ajustado MAE, MAPE, MSE, RMSE
# El coeficiente de determinación se puede ver como la varianza total explicada por el modelo, 
# es decir la proporción de la varianza en la variable dependiente que se puede predecir a partir de 
# las variables independientes
# Nos retorna el modelo entrenado y un diccionario con las principales métricas sobre el conjunto de
# entrenamiento y el de prueba

from sklearn.metrics import r2_score, mean_absolute_error, mean_absolute_percentage_error, mean_squared_error

def regression_metrics(X_train,X_test,y_train,y_test,y_pred_test):
    k_train = X_train.shape[1]
    n_train = len(X_train)
    k_test = X_test.shape[1]
    n_test = len(X_test)
    r2_test = r2_score(y_test, y_pred_test)
    adj_r2_test = 1-(1-r2_test)*(n_test-1)/(n_test-k_test-1)
    metrics = {"Test R^2": r2_test,
               "Test Adj R^2": adj_r2_test,
               "Test MAE": mean_absolute_error(y_test,y_pred_test),
               "Test MAPE": mean_absolute_percentage_error(y_test,y_pred_test),
               "Test RMSE": np.sqrt(mean_squared_error(y_test, y_pred_test))    
               }
    for item in metrics.items():
        print(item[0],"=",item[1])
    return metrics
    
def r2_metrics(X_train,X_test,y_train,y_test):
    k_train = X_train.shape[1]
    n_train = len(X_train)
    k_test = X_test.shape[1]
    n_test = len(X_test)
    r2_test = r2_score(y_test, y_pred_test)
    adj_r2_test = 1-(1-r2_test)*(n_test-1)/(n_test-k_test-1)
    return r2_test, adj_r2_test
```
